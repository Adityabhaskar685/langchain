```python
from langchain.embeddings import HuggingFaceEmbeddings
from llama_index import (
    SimpleDirectoryReader,
    VectorStoreIndex,
    ServiceContext,
)

from llama_index.llms import LlamaCPP
from llama_index.llms.llama_utils import messages_to_prompt, completion_to_prompt
```

```python
!wget https://huggingface.co/NousResearch/Nous-Capybara-7B-GGUF/resolve/main/ggml-model-Q5_K.gguf
```

```python
llm = LlamaCPP(
    model_path = './ggml-model-Q5_K.gguf',
    temperature = 0.1,
    max_new_tokens = 4000,
    context_window = 3900,
    #generate_kwargs = {},
    model_kwargs = {'n_gpu_layers': 43},
#     messages_to_prompt = messages_to_prompt,
#     completion_to_prompt = completion_to_prompt,
    verbose = True
    
)
```

```python
print(llm.complete("draft a legal document for selling a bike to another person.").text)
```

```python
from llama_index.prompts import PromptTemplate

template = (
    """
    USER: {question} \n\n
    ASSISTANT: \n
    """
)
qa_template = PromptTemplate(template)
```

```python
prompt = qa_template.format(question = "write a python program for deleting a node in binary search tree")
print(llm.complete(prompt).text)
```
